17.Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school.

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'School Code': ['S001', 'S002', 'S001', 'S003', 'S002'],
    'Age': [25, 24, 22, 23, 26]
}

df = pd.DataFrame(data)

# Group the DataFrame by 'School Code'
grouped = df.groupby('School Code')

# Calculate mean, min, and max age for each school
result = grouped['Age'].agg(['mean', 'min', 'max'])

# Rename the columns for clarity
result.columns = ['Mean Age', 'Min Age', 'Max Age']

# Display the result
print(result)
```

18.Write a Pandas program to split the following given dataframe into groups based on school code and class.

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'School Code': ['S001', 'S002', 'S001', 'S003', 'S002'],
    'Class': ['A', 'B', 'A', 'B', 'A'],
    'Age': [25, 24, 22, 23, 26]
}

df = pd.DataFrame(data)

# Group the DataFrame by 'School Code' and 'Class'
grouped = df.groupby(['School Code', 'Class'])

# Display the groups
for group_name, group_data in grouped:
    print(f"Group: {group_name}")
    print(group_data)
    print()
```

20.Write a Pandas program to find the index of a given substring of a DataFrame column.

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Text': ['Hello, this is a sample text.', 'This is another example.', 'Some more text.']
}

df = pd.DataFrame(data)

# Define the substring to search for
substring_to_find = 'example'

# Find the index of the substring in the 'Text' column
indices = df['Text'].str.find(substring_to_find)

# Display the indices
print(indices)
```

21.Write a Pandas program to swap the cases of a specified character column in a given DataFrame.

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Text': ['Hello World', 'Pandas is Great', 'SwAp CaSe ExAmPlE']
}

df = pd.DataFrame(data)

# Specify the character column to swap cases
column_to_swap = 'Text'

# Swap the cases of the specified column
df[column_to_swap] = df[column_to_swap].str.swapcase()

# Display the DataFrame with swapped cases
print(df)
```

23.Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1


```python
import matplotlib.pyplot as plt

# Read data from the text file
with open('test.txt', 'r') as file:
    lines = file.readlines()
    data = [line.strip().split() for line in lines]

x = [float(point[0]) for point in data]
y = [float(point[1]) for point in data]

# Create the line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot with Labels and Title')

# Show the plot
plt.show()
```

24.Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993

10-07-16,779.659973,779.659973,770.75,775.080017


```python
import matplotlib.pyplot as plt

# Sample financial data
dates = ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16']
opens = [774.25, 776.030029, 779.309998, 779.0, 779.659973]
highs = [776.065002, 778.710022, 782.070007, 780.47998, 779.659973]
lows = [769.5, 772.890015, 775.650024, 775.539978, 770.75]
closes = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

# Create line charts for Open, High, Low, and Close
plt.figure(figsize=(10, 6))
plt.plot(dates, opens, marker='o', label='Open', linestyle='-', color='b')
plt.plot(dates, highs, marker='o', label='High', linestyle='-', color='g')
plt.plot(dates, lows, marker='o', label='Low', linestyle='-', color='r')
plt.plot(dates, closes, marker='o', label='Close', linestyle='-', color='c')

plt.title('Alphabet Inc. Stock Prices (Oct 3, 2016 - Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
