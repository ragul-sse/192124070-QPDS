25.Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]
y2 = [8, 11, 9, 13, 16]

# Create line plots with legends, widths, and colors
plt.figure(figsize=(8, 6))

# Plot the first line with a label, width, and color
plt.plot(x, y1, label='Line 1', linewidth=2, color='blue')

# Plot the second line with a label, width, and color
plt.plot(x, y2, label='Line 2', linewidth=3, color='red')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines with Legends, Widths, and Colors')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

26.Write a Python program to create multiple plots.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with multiple subplots
plt.figure(figsize=(10, 6))

# Subplot 1: Sine curve
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Sine Curve')

# Subplot 2: Cosine curve
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Cosine Curve')

# Subplot 3: Tangent curve
plt.subplot(2, 1, 2)
plt.plot(x, y3)
plt.title('Tangent Curve')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

27.Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
import matplotlib.pyplot as plt

# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(languages, popularity, color='skyblue')

# Add labels and title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

30.Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]

# Define the width of the bars
bar_width = 0.35

# Create X values for men and women
x_men = np.arange(len(groups))
x_women = [x + bar_width for x in x_men]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(x_men, means_men, bar_width, label='Men', color='skyblue')
plt.bar(x_women, means_women, bar_width, label='Women', color='lightcoral')

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks([x + bar_width/2 for x in x_men], groups)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

31.Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
std_dev_men = [4, 3, 4, 1, 5]
std_dev_women = [3, 5, 2, 3, 3]

# Create X values for the groups
x = np.arange(len(groups))

# Create the stacked bar plot with error bars
plt.figure(figsize=(10, 6))
plt.bar(x, means_men, yerr=std_dev_men, capsize=6, label='Men', color='skyblue', alpha=0.7)
plt.bar(x, means_women, yerr=std_dev_women, capsize=6, label='Women', color='lightcoral', alpha=0.7, bottom=means_men)

# Add labels and title
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Stacked Bar Plot with Error Bars')

# Add a legend
plt.legend()

# Show the plot
plt.xticks(x, groups)
plt.tight_layout()
plt.show()
```

32.Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for X and Y
np.random.seed(0)  # Set a random seed for reproducibility
x = np.random.rand(50)  # 50 random X values
y = np.random.rand(50)  # 50 random Y values

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', marker='o', label='Random Data')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Graph with Random Data')

# Show the plot
plt.tight_layout()
plt.show()
```
